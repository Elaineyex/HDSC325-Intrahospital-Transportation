```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
waittime_data = [['Mean Baseline',221.63, 14.89, 11.48, 0],
                 ['Linear Regression',159.82, 12.64, 9.44, 0.28],
                 ['LASSO',159.42, 12.63, 9.44, 0.28],
                 ['Ridge', 159.41, 12.63, 9.44, 0.28],
                 ['Elastic Net', 159.74, 12.64, 9.47, 0.28],
                 ['KNN', 137.11, 11.71, 8.68, 0.38],
                 ['Decision Tree', 153.42, 12.39, 9.34, 0.31],
                 ['Random Forest', 126.25, 11.23, 8.4, 0.43],
                 ['XGBoost', 122.97, 11.09, 8.27, 0.45]]
waittime_df = pd.DataFrame(waittime_data, columns=['Model','MSE', 'RMSE', 'MAE','R^2'])
```

```python
completiontime_data = [['Mean Baseline', 32.60, 5.71, 4.5, 0],
                 ['Linear Regression',26.73, 5.17, 3.96, 0.18],
                 ['LASSO',26.73, 5.17, 3.96, 0.18],
                 ['Ridge', 26.73, 5.17, 3.96, 0.18],
                 ['Elastic Net', 26.77, 5.17, 3.97, 0.18],
                 ['KNN', 28.86, 5.37, 4.13, 0.11],
                 ['Decision Tree', 26.99, 5.19, 3.98, 0.17],
                 ['Random Forest', 26.41, 5.14, 3.94, 0.19],
                 ['XGBoost', 26.28, 5.13, 3.92, 0.19]]
completiontime_df = pd.DataFrame(completiontime_data, columns=['Model','MSE', 'RMSE', 'MAE','R^2'])
```

```python
totaltime_data = [['Mean Baseline', 230.24, 15.17, 11.72, 0.00],
                 ['Linear Regression',175.41, 13.24, 9.99, 0.24],
                 ['LASSO',175.40, 13.29, 10.00, 0.24],
                 ['Ridge', 175.40, 13.24, 9.99, 0.24],
                 ['Elastic Net', 175.60, 13.25, 10.00, 0.24],
                 ['KNN', 153.96, 12.41, 9.32, 0.33],
                 ['Decision Tree', 168.7, 12.99, 9.84, 0.27],
                 ['Random Forest', 141.86, 11.91, 9.02, 0.38],
                 ['XGBoost', 138.67, 11.78, 8.90, 0.40]]
totaltime_df = pd.DataFrame(totaltime_data, columns=['Model','MSE', 'RMSE', 'MAE','R^2'])
```

```python
cols = ['RMSE', 'MAE', 'R^2']

for col in cols:
    plt.figure(figsize = (8,6))
    plt.plot(waittime_df['Model'], waittime_df[col], '.-', label = 'wait time')
    plt.plot(completiontime_df['Model'], completiontime_df[col], '.-', label = 'completion time')
    plt.plot(totaltime_df['Model'], totaltime_df[col], '.-', label = 'total time')
    plt.xticks(rotation = 45)
    plt.legend()
    plt.ylabel(col)
    plt.title('Comparison of '+col+' Across Models')
    plt.show()
```
